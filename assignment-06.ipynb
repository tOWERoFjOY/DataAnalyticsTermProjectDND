# Question 6
## Improve the previous joint chart to be kernel density estimation (KDE) joint chart for:
* Homeruns vs strikeOuts for NYA 
* Homeruns vs strikeOuts for BOS 

```python
from scipy import stats
```

```python
# Your code goes here. Your chart for NYA should look like:
nya_year_df = df[(df.teamID == 'NYA')].pivot_table(index='yearID', aggfunc='sum')
xmin = nya_year_df.Homeruns.min()
xmax = nya_year_df.Homeruns.max()
ymin = nya_year_df.strikeOuts.min()
ymax = nya_year_df.strikeOuts.max()
X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([nya_year_df.Homeruns, nya_year_df.strikeOuts])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, X.shape)
```

```python
import matplotlib.pyplot as plt
print("Plot df Homeruns vs Strikeouts for NYA as a kernel density estimation.")
fig, ax = plt.subplots()
ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r,
        extent=[xmin, xmax, ymin, ymax])
ax.plot(nya_year_df.Homeruns, nya_year_df.strikeOuts, 'k.', markersize=3)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
plt.show()
```

```python
# Your code goes here. Your chart for BOS should look like:
bos_year_df = df[(df.teamID == 'BOS')].pivot_table(index='yearID', aggfunc='sum')
xmin = bos_year_df.Homeruns.min()
xmax = bos_year_df.Homeruns.max()
ymin = bos_year_df.strikeOuts.min()
ymax = bos_year_df.strikeOuts.max()
X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([bos_year_df.Homeruns, bos_year_df.strikeOuts])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, X.shape)
print("Plot df Homeruns vs Strikeouts for BOS as a kernel density estimation.")
```
